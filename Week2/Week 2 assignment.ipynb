# Bar Plot

## Next I plotted the number parks/recreation spaces by location type as a bar plot. In order to do so, I had to define a new variable, location_type. I created a bar plot after testing the regular plot. In order to create a bar plot, I defined my x and y axes and added a title to the bar plot.

```python
location_type=parks['LocationType'].value_counts()
```

```python
type(location_type)
```

```python
location_type.plot()
```

```python
location_type.plot.bar(x = 'Location Type', y = 'Count', title = 'Number of Parks and Recreation Facilities in LA')
```

## Above is my bar plot that displays the number of parks and recreation facilities in LA by location type. The bar plot shows that there are over 250 parks in LA.

# Trim data

## Next I opted to trim my data a bit. I chose to take out website, phone number, and geometry. I did this by first displaying the columns I wanted to keep and then defining a new variable, desired_columns. After this, I ran a sample to make sure I trimmed the data correctly and preserved my new dataframe by creating a copy of the data under a new name: parks_trimmed.

```python
parks[['LocationType', 'LocationName','Address','CouncilDistrict','GeoLat','GeoLong']]
```

```python
desired_columns=['LocationType','LocationName','Address','CouncilDistrict','GeoLat','GeoLong']
```

```python
parks[desired_columns].sample(5)
```

```python
parks_trimmed=parks[desired_columns].copy()
```

```python
parks_trimmed
```

## This last table above is my trimmed dataset, with location type, location name, address, council district, latitude, and longitude.

# Queries
